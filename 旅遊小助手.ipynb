# 旅遊小助手 app 設定

```python
title = "旅遊規劃小助手!"
system = "你是一個非常可靠的對話機器人，根據使用者想前往的旅遊地點以及旅遊方式，安排適合且完整的旅遊規劃，並請用台灣習慣的中文來回應。"
description = "你好，我是你的 AI 旅遊規劃小幫手, 想安排一場怎樣的旅行都可以問我哦 :)" #會顯示給使用者
model = "gemma3:4b"
```

**重點說明**
可能是因為一開始就使用了4b的版本 所以我滿滿意回答的結果，不只包含每天大致的行程，還列出了活動與住宿等的預算，還有行程活動可以調整的方向，所以我並沒有再改它的system
旅遊是現代人不可或缺的重要活動 所以我覺得旅遊安排小助手 十分實用 可以先透過ai了解我有甚麼樣的活動可以安排， 並進一步選擇想要的旅遊步調，旅遊方式，行程偏好等等

```python
initial_messages = [{"role":"system",
             "content":system},
            {"role":"assistant",
            'content':description}]
```

```python
def pipi(prompt, messages):
    messages.append({"role": "user", "content": prompt})
    chat_completion = client.chat.completions.create(
        messages=messages,
        model=model,
        )
    reply = chat_completion.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})
    #history = history + [[prompt, reply]]
    return messages, messages
```

```python
chatbot = gr.Chatbot(type="messages")
```

```python
with gr.Blocks(title=title) as demo:
    gr.Markdown(f"## 🤖 {title}\n{description}") ##調整字的大小 使用markdown來呈現
    chatbot = gr.Chatbot(type="messages")
    msg = gr.Textbox(label="輸入訊息")
    state = gr.State(initial_messages.copy())  # 務必用 copy() 歷史對話紀錄

    msg.submit(fn=pipi, inputs=[msg, state], outputs=[chatbot, state])

demo.launch(share=True, debug=True) #上架
```
